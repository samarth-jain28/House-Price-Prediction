```python
print(df['Sale Price'].mean())
print(df["Condition of the House"].unique())
print(df["Flat Area (in Sqft)"].std())
```

```python
plt.plot(df['Sale Price'].head(10), color='green', marker="o", markerfacecolor="blue", markersize=5, linewidth=1)
```

```python
df1 = df.groupby('Condition of the House')['ID'].count()
ar = np.array(df1)
plt.pie(ar)
plt.show()
```

```python
plt.scatter(x = df['Flat Area (in Sqft)'], y = df['Sale Price'], color = 'red')
plt.xlabel("Area of house")
plt.ylabel("Sale Price of house")
```

```python
plt.scatter(x = df['No of Bedrooms'], y = df['Sale Price'])
```

```python
df['Condition Sale']=0
for i in df['Condition of the House'].unique():
    df['Condition Sale'][df['Condition of the House'] == str(i)] = df['Sale Price'][df['Condition of the House'] == str(i) ].mean()
```

```python
plt.figure(dpi = 100)
plt.bar(df['Condition of the House'].unique(), df['Condition Sale'].unique())
plt.show()
```
